```python
#agrupo por agente y por llamadas recibidas, sumo, ordeno ascendecntemente y doy un index
llamadasAgenteIn = base.groupby('agente')['llamadas_in'].sum().sort_values().reset_index()
llamadasAgenteIn
```

```python
#agrupo por agente y por llamadas realizadas, sumo, ordeno ascendecntemente y doy un index
llamadasAgenteOut = base.groupby('agente')['llamadas_out'].sum().sort_values().reset_index()
llamadasAgenteOut
```

```python
figure,axis = plt.subplots(1,2,figsize=(20,7))
sns.barplot(x='agente', y='llamadas_in',data=llamadasAgenteIn,ax=axis[0])
axis[0].set_xticklabels(axis[0].get_xticklabels(),rotation = 80)
axis[0].set_title('Total Llamadas recibidas por agente')
sns.barplot(x='agente',y='llamadas_out',data=llamadasAgenteOut,ax=axis[1])
axis[1].set_xticklabels(axis[1].get_xticklabels(),rotation = 80)
axis[1].set_title(' Total llamadas realizadas por agente')
plt.show()
```
